```python
import leafmap
from samgeo import tms_to_geotiff
from samgeo.text_sam import LangSAM
```

```python
m = leafmap.Map(center=[34.040984, -118.491668], zoom=19, height="600px")
m.add_basemap("SATELLITE")
m
```

```python
image = "data/NAIP21QQ/03_quartered_tiles/tifs/m_4109344_nw_15_060_20210818_BL.tif"
m.layers[-1].visible = False
m.add_raster(image, layer_name="Image")
m
```

```python
sam = LangSAM()
```

```python
text_prompt = "buildings in the image"
```

```python
sam.predict(image, text_prompt, box_threshold=0.24, text_threshold=0.24)
```

```python
sam.show_anns(
    cmap="Blues",
    box_color="red",
    title="Automatic Segmentation of Barns in NAIP image",
    blend=True,
)
```

```python
sam.show_anns(
    cmap="Greens",
    add_boxes=False,
    alpha=0.5,
    title="Automatic Segmentation of Trees",
)
```
