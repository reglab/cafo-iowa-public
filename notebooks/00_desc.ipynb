```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

from munch import munchify
from cafo_iowa.utils.strata import generate_stratified_sample
```

```python
# load config
with open("cafo_iowa/config_old.yaml", "r") as file:
    config = munchify(yaml.safe_load(file))
```

```python
df = pd.read_csv("data/permits/processed/permits.csv")
naip_shp = gpd.read_feather("data/NAIP21QQ_shp/processed/03_quartered_tiles.feather")
```

```python
# subset dataset to only include permits with at least one swine
df = df[df["swine_animal_units"] > 0]
df.shape
```

```python
# histogram of df animal units (seaborn), capped at xlim 6000
plt.figure(figsize=(10, 6))
sns.histplot(df["animal_units"], bins=200)
plt.title("Histogram of Animal Units across permits")
plt.xlabel("Animal Units")
plt.ylabel("Frequency")
plt.xlim(0, 6000)
plt.savefig(
    config.plots.output_path + "histogram_max6000AU.png",
    dpi=config.plots.dpi,
    bbox_inches=config.plots.bbox_inches,
    pad_inches=0,
)
```
